```python
%reset -s -f
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('nnforces.out', 'r')
lines = file.readlines()
file.close()
```

```python
block = lines[18:]
block
```

```python
fx = []
fy = []
fz = []
for line in block:
    line_split = line.split()
    fx.append(float(line_split[0]))
    fy.append(float(line_split[1]))
    fz.append(float(line_split[2]))
    
fx = np.array(fx, dtype=np.float32)
fy = np.array(fy, dtype=np.float32)
fz = np.array(fz, dtype=np.float32)
```

```python
np.arange(0, fx.shape[0], 1)
```

```python
fig,ax = plt.subplots(3,1)

ax[0].plot(np.arange(0, fx.shape[0], 1), fx)
ax[1].plot(np.arange(0, fy.shape[0], 1), fy)
ax[2].plot(np.arange(0, fz.shape[0], 1), fz)


plt.show()
```
